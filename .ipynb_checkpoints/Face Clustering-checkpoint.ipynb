```python
base_dir = '/Users/babasarala/Desktop/final_face_clustering'
user_id = 8
threshold = 75
cnn_codes_dir = '%s/%i_prod_fb_tags_%i_cnn_codes'%(base_dir, user_id, threshold)
```

```python
img_urls, X = pu.combine_checkpoint_intermediate_results(cnn_codes_dir)
```

# Face Clustering

Use t-SNE + dbscan and visualize clusters

```python
# Apply t-SNE
n_components = 4
init = 'pca'
tsne = manifold.TSNE(n_components=n_components, init=init, random_state=0)
Y = tsne.fit_transform(X)
```

```python
import pickle as pkl
import os
if not os.path.isdir('%s/clustering/tSNE_%i'%(base_dir, n_components)):
    os.makedirs('%s/clustering/tSNE_%i'%(base_dir, n_components))
pkl.dump((Y, tsne), open('%s/clustering/tSNE_%i/%i_%i_model.p'%(base_dir, n_components, user_id, threshold), 'wb'))
```

```python
# visualization of clusters
Y = StandardScaler().fit_transform(Y)
plt.scatter(Y[:,0], Y[:,1])
plt.show()
```

```python
# collect some stats to figure out 
tree = cKDTree(Y)
d = tree.query(Y, 5)
dists = d[0]
neighbors = d[1]
```

```python
plt.hist(dists[:,4], bins=50)
plt.show()
```

```python
print np.min(dists[:,4])
print np.max(dists[:,4])
print np.median(dists[:,4])
```

```python
# Apply dbscan
min_samples = 5
eps = 90
db = DBSCAN(eps=eps, min_samples=min_samples).fit(Y)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
import pandas as pd
csv_savepath = '%s/clustering/tSNE_%i/%i_%i_%i_%i_clusters.csv' \
                %(base_dir, n_components, user_id, threshold, min_samples, eps)
df = pd.DataFrame({'img_url':img_urls, 'cluster_id':labels})
df.to_csv(csv_savepath, index=False)
```

```python
# Plot for 2D-points
colors = plt.cm.Spectral(np.linspace(0, 1, n_clusters_))
for cluster_idx, color in zip(range(n_clusters_), colors):
    idxs = np.where(labels == cluster_idx)[0]
    plt.scatter(Y[idxs,0], Y[idxs,1], c=color)
plt.show()
```

```python
cluster_map = {}
for k in range(n_clusters_):
    cluster_idxs = np.where(labels == k)[0]
    cluster_map[k] = [img_urls[c] for c in cluster_idxs]

num_img_per_cluster = []
for cluster_idx, img_list in cluster_map.iteritems():
    num_img_per_cluster.append(len(img_list))    
```
