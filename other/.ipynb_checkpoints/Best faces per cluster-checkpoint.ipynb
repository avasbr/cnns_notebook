```python
def filter_cluster_by_size(cluster_df, min_size=10, max_size=50):
    assert 'cluster_idx' in cluster_df
    cluster_counts = cluster_df[['cluster_idx', 'img_url']] \
                               .groupby('cluster_idx') \
                               .count() \
                               .reset_index() \
                               .rename(columns={'img_url':'count'})
    filtered_cluster_counts = cluster_counts[(cluster_counts['count'] >= min_size) &
                                             (cluster_counts['count'] <= max_size)]
    cols_to_keep = list(cluster_df.columns)
    return pd.merge(cluster_df, filtered_cluster_counts, on='cluster_idx')[cols_to_keep]

def compute_variance(X, dist_type='euclidean'):
    if dist_type == 'euclidean':
        X_mu = np.mean(X, axis=0)
        return np.mean(np.sum((X - X_mu)**2, axis=1))
    elif dist_type == 'cosine':
        pass
    
def compute_within_cluster_variance(cluster_ids, cluster_df, feat_dim, type='euclidean'):
    feat_cols = ['f%i' % i for i in range(feat_dim)]
    within_cluster_variance = 0
    for cluster_id in cluster_ids:
        pts = cluster_df[cluster_df['cluster_idx'] == cluster_id][feat_cols].values
        within_cluster_variance += compute_variance(pts)
    return within_cluster_variance

def compute_between_cluster_variance(cluster_ids, cluster_df, feat_dim):
    feat_cols = ['f%i' % i for i in range(feat_dim)]
    between_cluster_variance = 0
    centroids = np.empty((len(cluster_ids), feat_dim))
    for idx, cluster_id in enumerate(cluster_ids):
        pts = cluster_df[cluster_df['cluster_idx'] == cluster_id][feat_cols].values
        centroids[idx] = np.mean(pts, axis=0)
        
    between_cluster_variance = compute_variance(centroids)
    return between_cluster_variance

def compute_fisher_criterion(cluster_ids, cluster_df, feat_dim):
    S_w = compute_within_cluster_variance(cluster_ids, cluster_df, feat_dim)
    S_b = compute_between_cluster_variance(cluster_ids, cluster_df, feat_dim)
    if S_w == 0:
        return np.inf
    return S_b/S_w

def compute_cluster_stats(cluster_df, feat_dim):
    
    feat_cols = ['f%i'%i for i in range(feat_dim)]
    assert 'cluster_idx' in cluster_df
    
    cluster_ids = cluster_df['cluster_idx'].unique()
    cluster_inv_var = [None]*len(cluster_ids)
    cluster_freq = [None]*len(cluster_ids)
    for idx, cluster_id in enumerate(cluster_ids):
        pts = cluster_df[cluster_df['cluster_idx'] == cluster_id][feat_cols].values
        var = compute_variance(pts)
        if var == 0:
            cluster_inv_var[idx] = 0
        else:
            cluster_inv_var[idx] = 1./compute_variance(pts)
        cluster_freq[idx] = len(pts)
        
    cluster_stats = pd.DataFrame({'cluster_idx':cluster_ids,
                                  'inv_var':cluster_inv_var,
                                  'freq':cluster_freq})
    cluster_stats['norm_inv_var'] = cluster_stats['inv_var']/cluster_stats['inv_var'].max()
    cluster_stats['norm_freq'] = cluster_stats['freq']/cluster_stats['freq'].max()
    
    return cluster_stats

def compute_simplex_score(lambda_vals, vals):
    assert np.sum(lambda_vals) == 1.0
    return np.dot(lambda_vals, vals)



#cluster_stats['score'] = lmbda*cluster_stats['norm_inv_var'] + (1.-lmbda)*cluster_stats['norm_freq']
#cluster_stats[cluster_stats['freq'] >= min_freq].sort(columns=['score'], ascending=False, inplace=True)
#best_clusters_stats = cluster_stats.sort(columns=['score'], ascending=False).head()
        

def compute_within_cluster_and_length_metric(cluster_id, cluster_df, feat_dim, w=0.2):
    assert w <= 1.0
    feat_cols = ['f%i' % i for i in range(feat_dim)]
    pts = cluster_df[cluster_df['cluster_idx'] == cluster_id][feat_cols].values
    S_w = compute_variance(pts)
    f = len(pts)
    return w*f + (1-w)*S_w
    

def sort_cluster_ids_and_get(fcn, cluster_df, feat_dim, num_clusters=15):
    cluster_list = cluster_df['cluster_idx'].unique()
    if num_clusters is None:
        num_clusters = len(cluster_list)
    f_values = []
    for cluster_id in cluster_list:
        f_values.append(fcn([cluster_id], cluster_df, feat_dim))
    idxs = np.argsort(f_values)
    sorted_clusters = [cluster_list[idx] for idx in idxs]
    return sorted_clusters[:num_clusters]
    
def expose_best_clusters(obj_fcn, cluster_df, feat_dim, num_clusters=7, cluster_list=None):
    if cluster_list is None:
        cluster_list = cluster_df['cluster_idx'].unique()
    best_cluster_ids = []
    best_obj_value = 0
    print 'Number of subsets to test (brute force): %i'%(nCk(len(cluster_list), num_clusters))
    for cluster_ids in combinations(cluster_list, num_clusters):
        curr_obj_value = obj_fcn(cluster_ids, cluster_df, feat_dim)
        if curr_obj_value > best_obj_value:
            best_obj_value = curr_obj_value
            best_cluster_ids = cluster_ids
    return best_cluster_ids

def nCk(n, k):
    f = math.factorial
    return f(n) / f(k) / f(n-k)
```
